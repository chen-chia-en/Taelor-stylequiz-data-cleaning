```python
# delete multiple choices ex: others/ Search engine (Google, Yahoo, etc.) => Search engine (Google, Yahoo, etc.)
clean_stylequiz = pd.read_csv('clean_stylequiz_taelor.csv')
```

```python
clean_stylequiz = clean_stylequiz.drop('Unnamed: 0', axis = 1)
clean_stylequiz.head(5)
```

```python
clean_stylequiz.iloc[clean_stylequiz['acquisition source'] == 'News/ Blog/ Online articles ',:] = 'News/ Blog/ Online articles'
```

```python
clean_stylequiz.iloc[clean_stylequiz['acquisition source'] == 'Facebook / Friend/ Family',:] = 'socail_media'
clean_stylequiz['acquisition source'].value_counts()
```

```python
# clean_stylequiz['acquisition source'].value_counts()
clean_stylequiz['acquisition source'].unique()
```

```python
#因為社群媒體 FB Titter clubhouse email人數少 合併成 socail_media
clean_stylequiz.iloc[(clean_stylequiz['acquisition source'] == 'Email') | 
                     (clean_stylequiz['acquisition source'] == 'Facebook') |
                     (clean_stylequiz['acquisition source'] == 'Twitter') |
                     (clean_stylequiz['acquisition source'] == 'Clubhouse') ,:] = 'socail_media'
```
